```python
import numpy as np
import matplotlib.pyplot as plot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline 
```

```python
dataset1 = np.loadtxt("poblacionCTG.txt",delimiter=";")
x= dataset1[:,1]
y= dataset1[:,0]
x = x.reshape(-1,1)
y = y.reshape(-1,1)

dataset2 = np.loadtxt("poblacionBogota.txt",delimiter=";")
a= dataset2[:,1]
b= dataset2[:,0]
a = a.reshape(-1,1)
b = b.reshape(-1,1)

dataset3 = np.loadtxt("poblacionMedellin.txt",delimiter=";")
c= dataset3[:,1]
d= dataset3[:,0]
c = c.reshape(-1,1)
d = d.reshape(-1,1)

dataset4 = np.loadtxt("poblacionBquilla.txt",delimiter=";")
e= dataset4[:,1]
f= dataset4[:,0]
e = e.reshape(-1,1)
f = f.reshape(-1,1)
```

```python
#Dividiendo el dataset para entrenar y evaluar
#1/3 para test_zise (pruebas)
#2/3 para entrenar con los datos
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1/3, random_state = 0)
linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)
yPrediction = linearRegressor.predict(xTest)
```

```python
h = int(input("Introduce el año:"))
ynew = linearRegressor.predict([[h]])
#round es para aproximar 
ynew = round(float(ynew))
```

```python


print("_______________________________________________________\n")
print("La población de CARTAGENA para el año",h, "es de: ",ynew)
plot.scatter(xTrain, yTrain, color = 'red')
plot.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plot.title('Población vs Año (Conjunto de entrenamiento)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()

# Visualising the test set results
plot.scatter(xTest, yTest, color = 'red')
plot.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plot.title('Población vs Año (Equipo de prueba)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()
```

```python
#--------------------> BOGOTA  <----------------------------------------------------
#Dividiendo el dataset para entrenar y evaluar
#1/3 para test_zise (pruebas)
#2/3 para entrenar con los datos
aTrain, aTest, bTrain, bTest = train_test_split(a, b, test_size = 1/3, random_state = 0)
linearRegressor = LinearRegression()
linearRegressor.fit(aTrain, bTrain)
bPrediction = linearRegressor.predict(aTest)
bnew = linearRegressor.predict([[h]])
#round es para aproximar 
bnew = round(float(bnew)) 
print("_______________________________________________________\n")
print("La población de BOGOTÁ para el año",h, "es de: ",bnew)
# Visualising the training set results
plot.scatter(aTrain, bTrain, color = 'red')
plot.plot(aTrain, linearRegressor.predict(aTrain), color = 'blue')
plot.title('Población vs Año (Conjunto de entrenamiento)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()
# Visualising the test set results
plot.scatter(aTest, bTest, color = 'red')
plot.plot(aTrain, linearRegressor.predict(aTrain), color = 'blue')
plot.title('Población vs Año (Equipo de prueba)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()
```

```python
#--------------------> MEDELLÍN  <----------------------------------------------------
#Dividiendo el dataset para entrenar y evaluar
#1/3 para test_zise (pruebas)
#2/3 para entrenar con los datos
cTrain, cTest, dTrain, dTest = train_test_split(c, d, test_size = 1/5, random_state = 0)
linearRegressor = LinearRegression()
linearRegressor.fit(cTrain, dTrain)
dPrediction = linearRegressor.predict(dTest)
dnew = linearRegressor.predict([[h]])
#round es para aproximar 
dnew = round(float(dnew)) 
print("_______________________________________________________\n")
print("La población de MEDELLÍN para el año",h, "es de: ",dnew)
# Visualising the training set results
plot.scatter(cTrain, dTrain, color = 'blue')
plot.plot(cTrain, linearRegressor.predict(cTrain), color = 'red')
plot.title('Población vs Año (Conjunto de entrenamiento)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()

# Visualising the test set results
plot.scatter(cTest, dTest, color = 'blue')
plot.plot(cTrain, linearRegressor.predict(cTrain), color = 'red')
plot.title('Población vs Año (Equipo de prueba)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()
```

```python
#--------------------> BARRANQUILLA  <----------------------------------------------------
#Dividiendo el dataset para entrenar y evaluar
#1/3 para test_zise (pruebas)
#2/3 para entrenar con los datos
eTrain, eTest, fTrain, fTest = train_test_split(e, f, test_size = 1/3, random_state = 0)
linearRegressor = LinearRegression()
linearRegressor.fit(eTrain, fTrain)
fPrediction = linearRegressor.predict(fTest)
fnew = linearRegressor.predict([[h]])
#round es para aproximar 
fnew = round(float(fnew)) 
print("_______________________________________________________\n")
print("La población de BARRANQUILLA para el año",h, "es de: ",fnew)
# Visualising the training set results
plot.scatter(eTrain, fTrain, color = 'red')
plot.plot(eTrain, linearRegressor.predict(eTrain), color = 'green')
plot.title('Población vs Año (Conjunto de entrenamiento)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()

# Visualising the test set results
plot.scatter(eTest, fTest, color = 'red')
plot.plot(eTrain, linearRegressor.predict(eTrain), color = 'green')
plot.title('Población vs Año (Equipo de prueba)')
plot.xlabel('Años')
plot.ylabel('Población')
plot.show()
```
